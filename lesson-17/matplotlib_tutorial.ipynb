## **Matplotlib**

*pip install matplotlib*

Matplotlib is a popular Python library used for creating static, interactive, and animated visualizations. It provides a flexible framework for producing a wide range of plots like line charts, bar graphs, scatter plots, and histograms. Its pyplot module offers a MATLAB-like interface, making it easy to create and customize plots with simple commands. Matplotlib is widely used in data analysis and scientific research for its versatility and integration with other libraries like NumPy and pandas.

### **Pyplot**

- Figures and subplots
- Colors, Markers and Line Styles
- Line plots
- Bar plots
- Histograms
- Here are some famous functions in the matplotlib.pyplot package:

1. **`plot()`** - Creates line plots.
2. **`scatter()`** - Creates scatter plots.
3. **`bar()`** - Creates bar charts.
4. **`hist()`** - Creates histograms.
5. **`imshow()`** - Displays images.
6. **`show()`** - Displays the plots.
7. **`xlabel()`** / ylabel() - Labels the x-axis and y-axis.
8. **`title()`** - Adds a title to the plot.
9. **`legend()`** - Adds a legend to the plot.
10. **`grid()`** - Adds grid lines to the plot.
11. **`xlim() / ylim()`** - Sets the limits for the x-axis and y-axis.

### **Pyplot Figures and Subplots**


- **`figure()`**
- **`subplot()`**

**Figures:** O'zi kattaroq grafik yoki figura yaratish uchun `figure()` funksiyasi ishlatiladi.

**Subplots:** Bir nechta kichik grafiklarni bir figuraning ichida joylashtirish uchun `subplot()` funksiyasidan foydalaniladi.

### **Colors, Markers and Line Styles**

**Colors:** Ranglarni tanlash uchun matn, HEX kodlari yoki RGB qiymatlari ishlatilishi mumkin.

**Markers:** Grafiklardagi nuqtalarni belgilash uchun ishlatiladi, masalan, `o`, `s`, `^`, `*` va boshqalar.

**Line Styles:** Chiziq uslublarini aniqlash uchun `-`, `--`, `-.`, `:` kabi belgilar ishlatiladi.

### **Line plots**

Chiziq diagrammalar oddiy ma'lumotlarni ko'rsatish uchun ishlatiladi.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
y = np.sin(x)

plt.plot(x, y, label='Sine wave', color='b', marker='o')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```

### **Bar plots**

Ustun diagrammalarni yaratish uchun `bar()` funksiyasi ishlatiladi.

```python
categories = ['A', 'B', 'C', 'D']
values = [3, 7, 5, 2]

plt.bar(categories, values, color='skyblue')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Plot Example')
plt.show()
```

### **Histograms**

Gistogrammalarni yaratish uchun `hist()` funksiyasi ishlatiladi.

```python
data = np.random.randn(1000)

plt.hist(data, bins=30, color='purple', edgecolor='black')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram Example')
plt.show()
```

### **Matplotlib pyplot paketida ishlatiladigan mashhur funksiyalar haqida qisqacha ma'lumot:**

 `plot()`: Chiziqli grafiklar yaratadi.

`scatter()`: Nuqta diagrammalar yaratadi.

`bar()`: Ustun diagrammalar yaratadi.

`hist()`: Gistogrammalar yaratadi.

`imshow()`: Rasmlarni ko'rsatadi.

`show()`: Grafiklarni ko'rsatadi.

`xlabel() / ylabel()`: X va Y o'qlarini belgilaydi.

`title()`: Grafikga sarlavha qo'shadi.

`legend()`: Grafikga tushuntirish (legenda) qo'shadi.

`grid()`: Grafikga tarmoq chiziqlarini qo'shadi.

`xlim() / ylim()`: X va Y o'qlarining chegaralarini belgilaydi.

```python
import matplotlib.pyplot as plt
import numpy as np

# Yangi figura va kichik diagrammalar yaratish
fig = plt.figure(figsize=(8, 6), dpi=80)

# Birinchi kichik diagramma
ax1 = fig.add_subplot(2, 1, 1)
x = np.linspace(0, 10, 100)
y = np.sin(x)
ax1.plot(x, y, label='Sine wave')
ax1.set_title('Kichik Diagramma 1')
ax1.legend()

# Ikkinchi kichik diagramma
ax2 = fig.add_subplot(2, 1, 2)
y2 = np.cos(x)
ax2.plot(x, y2, label='Cosine wave', color='r')
ax2.set_title('Kichik Diagramma 2')
ax2.legend()

# Sarlavha va qutqarish
fig.suptitle('Matplotlibda Figuralar va Kichik Diagrammalar')
plt.savefig('example_figure.png')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Yangi figura yaratish
fig = plt.figure(figsize=(8, 6))

# Birinchi kichik diagramma
ax1 = plt.subplot(2, 1, 1)
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
ax1.plot(x, y1, label='Sine wave')
ax1.set_title('Kichik Diagramma 1')
ax1.legend()

# Ikkinchi kichik diagramma
ax2 = plt.subplot(2, 1, 2)
y2 = np.cos(x)
ax2.plot(x, y2, label='Cosine wave', color='r')
ax2.set_title('Kichik Diagramma 2')
ax2.legend()

# Figuraning sarlavhasi va ko'rsatish
fig.suptitle('Matplotlibda Kichik Diagrammalar')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Yangi figura va kichik diagrammalar yaratish
fig, axes = plt.subplots(2, 1, figsize=(8, 6))

# Birinchi kichik diagramma
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
axes[0].plot(x, y1, label='Sine wave')
axes[0].set_title('Kichik Diagramma 1')
axes[0].legend()

# Ikkinchi kichik diagramma
y2 = np.cos(x)
axes[1].plot(x, y2, label='Cosine wave', color='r')
axes[1].set_title('Kichik Diagramma 2')
axes[1].legend()

# Figuraning sarlavhasi va ko'rsatish
fig.suptitle('Matplotlibda plt.subplots() yordamida Kichik Diagrammalar')
plt.show()

```

### **Markerlar**
**Markerlar grafiklardagi individual ma'lumot nuqtalarini ajratib ko'rsatadi.**

Marker	Tavsif
- `'.'`	Nuqta
- `','`	Piksel
- `'o'`	Doira
- `'v'`	Pastga Uchburchak
- `'^'`	Yuqoriga Uchburchak
- `'<'`	Chapga Uchburchak
- `'>'`	O'ngga Uchburchak
- `'s'`	Kvadrat
- `'*'`	Yulduzcha
- `'x'`	X harfi
- `'D'`	Romboid
- `'H'`	Geksagon

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]
z = [30, 10, 15, 50]

# Rang va markerlardan foydalanish
plt.plot(x, y, color='b', marker='o', label='Line y')
plt.plot(x, z, color='r', marker='x', label='Line z')
plt.xlabel('X o\'qi')
plt.ylabel('Y va Z o\'qi')
plt.title('Ranglar va Markerlar bilan Grafiklar')
plt.legend()
plt.show()
```

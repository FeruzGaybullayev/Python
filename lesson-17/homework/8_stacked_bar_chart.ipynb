#### **8. Stacked Bar Chart**
- **Task**: Create a stacked bar chart that shows the contribution of three different categories (`'Category A'`, `'Category B'`, and `'Category C'`) over four time periods (`'T1'`, `'T2'`, `'T3'`, `'T4'`). Use sample data for each category at each time period. Customize the chart with a title, axis labels, and a legend.
------------

```python
import matplotlib.pyplot as plt
import numpy as np

# Time periods and categories
time_periods = ['T1', 'T2', 'T3', 'T4']
category_a = [10, 15, 20, 25]
category_b = [20, 25, 15, 10]
category_c = [30, 20, 25, 30]

# Create the bar chart
fig, ax = plt.subplots()

# Plot data
ax.bar(time_periods, category_a, label='Category A', color='blue')
ax.bar(time_periods, category_b, label='Category B', bottom=category_a, color='green')
ax.bar(time_periods, category_c, label='Category C', bottom=np.array(category_a) + np.array(category_b), color='red')

# Add title and labels
ax.set_title('Stacked Bar Chart of Different Categories Over Time Periods')
ax.set_xlabel('Time Periods')
ax.set_ylabel('Values')

# Add legend
ax.legend()

# Show the plot
plt.show()
```

Task_1. Basic Plotting
Task: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.

$ f(x) = x^2 - 4x + 4 $ for $ x $ 

```python
# import matplotlib and numpy
import matplotlib.pyplot as plt
import numpy as np

# Funksiya yozish
def f_(x):
    return x**2 - 4*x + 4

x = np.linspace(-10, 10)
y = f_(x)

plt.plot(x, y)
plt.xlabel("X o'qi")
plt.ylabel("Y o'qi")
plt.title("$ Kvadrat $ $ funktsiya $")
```

Task_2. $ \sin(x) $ va $ \cos(x) $ ni 0 dan $ 2\pi $ gacha boʻlgan $ x $ qiymatlari uchun bir xil grafikda chizing. Ikki funktsiyani farqlash uchun turli xil chiziq uslublari, markerlar va ranglardan foydalaning. Legend qo'shing.

```python
# x qiymatlarini yaratish
x = np.linspace(0, 2*np.pi)

# y qiymatlarini yaratish
sin_x = np.sin(x)
cos_x = np.cos(x)

# Sarlavha va etiketkalar qo'shish
plt.plot(x, sin_x, color='green', marker='o', label='sin(x)')
plt.plot(x, cos_x, color='blue', label='cos(x)')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Grafikni ko'rsatish
plt.show()
```

Task_3. Create a 2x2 grid of subplots. In each subplot, plot:

Top-left: $ f(x) = x^3 $
Top-right: $ f(x) = \sin(x) $
Bottom-left: $ f(x) = e^x $
Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)
Customize each plot with titles, axis labels, and different colors.



```python
# X qiymatlari
x = np.linspace(-10, 10, 400)
x_ = np.linspace(0, 10, 400)

# Figure 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Yuqori chap: f(x) = x**3
axs[0, 0].plot(x, x**3, color='red')
axs[0, 0].set_title('$f(x) = x^3$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')

# Yuqori o'ng: f(x) = sin(x)
axs[0, 1].plot(x, np.sin(x), color='green')
axs[0, 1].set_title('$f(x) = sin(x)$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')

# Yuqori chap: f(x) = x**3
axs[1, 0].plot(x, x**3, color='red')
axs[1, 0].set_title(r'$f(x) = e^x$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')

# Yuqori o'ng: f(x) = sin(x)
axs[1, 1].plot(x_, np.sin(x_), color='green')
axs[1, 1].set_title(r'$f(x) = log(x+1)$, ($x \geq 0$ uchun)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')

# Sarlavha
fig.suptitle("2x2 o'lchamli diagrammalar jadvali", fontsize=16)

plt.tight_layout()
plt.show()
```

Task_4.  Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid

```python
# (0, 10) oraliqda tasodifiy 100 ta raqam tanlab x, y ga beramiz
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Ranglar va markerlar tanlash
colors = np.random.rand(100)
markers = np.random.choice(['o', 's', '^', 'D'], size=100)

# Scatter plot yaratish
plt.figure(figsize=(10, 6))
for i in range(100):
    plt.scatter(x[i], y[i], color=plt.cm.viridis(colors[i]), marker=markers[i])
plt.title('Scatter Plot of 100 Random Points')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.grid(True)
plt.show()
```

Task_5. Generate a random dataset of 1000 values sampled from a normal distribution (mean=0, std=1). Plot a histogram of the data with 30 bins. Add a title and axis labels. Adjust the transparency of the bars using the alpha parameter.

```python
# Normal taqsimotdan (mean=0, std=1)
data = np.random.normal(0, 1, 1000)

# Plot a hitogram
plt.figure(figsize=(10, 8))
plt.hist(data, bins=30, color='g', edgecolor='b', alpha=1)
plt.title('Histogram of normally distributed data')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.show()
```

Task_6. $ f(x, y) = \cos(x^2 + y^2) $ funksiyasi uchun $ x $ va $ y $ qiymatlari oralig'ida -5 dan 5 gacha bo'lgan 3D sirt grafigini yarating. mos rang xaritasi va rang paneli qo'shing. O'qlar va sarlavha uchun tegishli teglarni o'rnating.

```python
# X and Y
x = np.linspace(-5, 5, 1000)
y = np.linspace(-5, 5, 1000)
X, Y = np.meshgrid(x, y)

# Hisoblash
Z = np.cos(X**2 + Y**2)

# Grafik
fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')
surf = ax.plot_surface(X, Y, Z, cmap='viridis', edgecolor='none')
fig.colorbar(surf, shrink=0.5, aspect=5)

# Sarlavha
ax.set_title("$f(x, y) = \cos(x^2 + y^2)$ Funktsiyasining 3D Grafi")
ax.set_xlabel("X o'qi")
ax.set_ylabel("Y o'qi")
ax.set_zlabel("Z o'qi")

plt.show()
```

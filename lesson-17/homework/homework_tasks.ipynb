Task_1. Basic Plotting
Task: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.

$ f(x) = x^2 - 4x + 4 $ for $ x $ 

```python
# import matplotlib and numpy
import matplotlib.pyplot as plt
import numpy as np

# Funksiya yozish
def f_(x):
    return x**2 - 4*x + 4

x = np.linspace(-10, 10)
y = f_(x)

plt.plot(x, y)
plt.xlabel("X o'qi")
plt.ylabel("Y o'qi")
plt.title("$ Kvadrat $ $ funktsiya $")
```

Task_2. $ \sin(x) $ va $ \cos(x) $ ni 0 dan $ 2\pi $ gacha boʻlgan $ x $ qiymatlari uchun bir xil grafikda chizing. Ikki funktsiyani farqlash uchun turli xil chiziq uslublari, markerlar va ranglardan foydalaning. Legend qo'shing.

```python
# x qiymatlarini yaratish
x = np.linspace(0, 2*np.pi)

# y qiymatlarini yaratish
sin_x = np.sin(x)
cos_x = np.cos(x)

# Sarlavha va etiketkalar qo'shish
plt.plot(x, sin_x, color='green', marker='o', label='sin(x)')
plt.plot(x, cos_x, color='blue', label='cos(x)')
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()

# Grafikni ko'rsatish
plt.show()
```

Task_3. Create a 2x2 grid of subplots. In each subplot, plot:

Top-left: $ f(x) = x^3 $
Top-right: $ f(x) = \sin(x) $
Bottom-left: $ f(x) = e^x $
Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)
Customize each plot with titles, axis labels, and different colors.



```python
import matplotlib.pyplot as plt
import numpy as np

# X qiymatlari
x = np.linspace(-10, 10, 400)
x_ = np.linspace(0, 10, 400)

# Figure 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Yuqori chap: f(x) = x**3
axs[0, 0].plot(x, x**3, color='red')
axs[0, 0].set_title('$f(x) = x^3$')
axs[0, 0].set_xlabel('x')
axs[0, 0].set_ylabel('f(x)')

# Yuqori o'ng: f(x) = sin(x)
axs[0, 1].plot(x, np.sin(x), color='green')
axs[0, 1].set_title('$f(x) = sin(x)$')
axs[0, 1].set_xlabel('x')
axs[0, 1].set_ylabel('f(x)')

# Yuqori chap: f(x) = x**3
axs[1, 0].plot(x, x**3, color='red')
axs[1, 0].set_title('$f(x) = e^x$')
axs[1, 0].set_xlabel('x')
axs[1, 0].set_ylabel('f(x)')

# Yuqori o'ng: f(x) = sin(x)
axs[1, 1].plot(x_, np.sin(x_), color='green')
axs[1, 1].set_title('$f(x) = log(x+1)$, ($x \geq 0$ uchun)')
axs[1, 1].set_xlabel('x')
axs[1, 1].set_ylabel('f(x)')

# Sarlavha
fig.suptitle("2x2 o'lchamli diagrammalar jadvali", fontsize=16)

plt.tight_layout()
plt.show()
```

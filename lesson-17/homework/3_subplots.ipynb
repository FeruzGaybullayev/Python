#### **3. Subplots**
- **Task**: Create a 2x2 grid of subplots. In each subplot, plot:
  - Top-left: $ f(x) = x^3 $
  - Top-right: $ f(x) = \sin(x) $
  - Bottom-left: $ f(x) = e^x $
  - Bottom-right: $ f(x) = \log(x+1) $ (for $ x \geq 0 $)

  Customize each plot with titles, axis labels, and different colors.

---

```python
import matplotlib.pyplot as plt
import numpy as np

# Define range of X values
x = np.linspace(-10, 10, 400)
x_ = np.linspace(0, 10, 400)

# Figure 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

# Top left
axs[0, 0].plot(x, x**3, color='red')
axs[0, 0].set_title('$f(x) = x^3$')
axs[0, 0].set_xlabel('X')
axs[0, 0].set_ylabel('Y')

# Top right
axs[0, 1].plot(x, np.sin(x), color='green')
axs[0, 1].set_title('$f(x) = sin(x)$')
axs[0, 1].set_xlabel('X')
axs[0, 1].set_ylabel('Y')

# Bottom left 
axs[1, 0].plot(x, np.exp(x), color='blue')
axs[1, 0].set_title('$f(x) = e^x$')
axs[1, 0].set_xlabel('X')
axs[1, 0].set_ylabel('Y')

# Bottom right
axs[1, 1].plot(x, np.log(x+1), color='orange')
axs[1, 1].set_title(r'$f(x) = log(x+1)$, (for $x \geq 0$)')
axs[1, 1].set_xlabel('X')
axs[1, 1].set_ylabel('Y')


# Overall title 
fig.suptitle("2x2 Grid of Different Funcsions", fontsize=15)

plt.tight_layout()
plt.show()
```

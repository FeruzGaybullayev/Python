#### **1. Basic Plotting**
- **Task**: Plot the function $ f(x) = x^2 - 4x + 4 $ for $ x $ values between -10 and 10. Customize the plot with appropriate labels for the axes and a title.

---

```python
import matplotlib.pyplot as plt
import numpy as np

# Enter funcsion
def f_(x):
    return x**2 - 4*x + 4

# The values that X takes
x = np.linspace(-10, 10)

# Funcsion use for x
y = f_(x)

# Axes and titles
plt.plot(x, y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Quadratic funcsion')

# Drawing output
plt.show()
```

#### **4. Scatter Plot**
- **Task**: Create a scatter plot of 100 random points in a 2D space. The x and y values should be randomly generated from a uniform distribution between 0 and 10. Use different colors and markers for the points. Add a title, axis labels, and a grid.

---

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate 100 random points for x and y from a uniform distribution between 0 and 10
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)

# Create a scatter plot
plt.scatter(x, y, c=np.random.rand(100), marker='o', alpha=0.7, edgecolors='w', cmap='viridis')

# Add title and labels
plt.title('Scatter Plot of 100 Random Points')
plt.xlabel('X')
plt.ylabel('Y')

# Add grid
plt.grid(True)

# Show the plot
plt.show()
```

#### **2. Sine and Cosine Plot**
- **Task**: Plot $ \sin(x) $ and $ \cos(x) $ on the same graph for $ x $ values ranging from 0 to $ 2\pi $. Use different line styles, markers, and colors to distinguish between the two functions. Add a legend.

---

```python
import matplotlib.pyplot as plt
import numpy as np

# Assign a value to to X
x = np.linspace(0, 2*np.pi)

# Writing funcsions
def sinus(x):
    return np.sin(x)

def cosinus(x):
    return np.cos(x)

# Enter title and libel
plt.plot(x, sinus(x), linestyle='-', color="green", marker='o', label='sin(x)')
plt.plot(x, cosinus(x), linestyle='--', color='blue', marker='D', label='cos(x)')
plt.title("Plot of sin(x) and cos(x)")

# Show the plot
plt.grid(True)
plt.show()
```
